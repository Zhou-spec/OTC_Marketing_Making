```python
# download the npy file data

import numpy as np
import matplotlib.pyplot as plt
%matplotlib ipympl

# load the data
loss_mlp = np.load('loss_policy_mlp.npy')
print(loss_mlp.shape)
print(loss_mlp)
```

```python
#draw the figure
plt.figure()
plt.plot(loss_mlp[5:])
plt.xlabel('iteration')
plt.ylabel('loss')
plt.title('loss of policy mlp')
plt.show()
```

```python
loss_res_cnn = np.load('loss_policy_res_cnn_loss.npy')
print(loss_res_cnn.shape)
print(loss_res_cnn)

x = [i / loss_res_cnn[0] for i in loss_res_cnn] 
#draw the figure
plt.figure()
plt.plot(x)
plt.xlabel('iteration')
plt.ylabel('loss')
plt.title('loss of policy res_cnn')
plt.ylim(0, 5)
plt.show()
```

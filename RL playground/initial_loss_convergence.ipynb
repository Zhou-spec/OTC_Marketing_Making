```python
# download the npy file data

import numpy as np
import matplotlib.pyplot as plt
%matplotlib ipympl

# load the data
loss_mlp = np.load('loss_policy_mlp.npy')
print(loss_mlp.shape)
print(loss_mlp)
```

```python
#draw the figure
x_mlp = [i / loss_mlp[0] for i in loss_mlp]
plt.figure()
plt.plot(x_mlp)
plt.xlabel('iteration')
plt.ylabel('loss')
plt.title('loss of policy mlp')
plt.ylim(0, 20)
plt.savefig('loss_policy_mlp.png')
plt.show()
```

```python
loss_res_cnn = np.load('loss_policy_res_cnn_loss.npy')
print(loss_res_cnn.shape)
print(loss_res_cnn)

x = [i / loss_res_cnn[0] for i in loss_res_cnn] 
#draw the figure
plt.figure()
plt.plot(x)
plt.xlabel('iteration')
plt.ylabel('loss')
plt.title('loss of policy res_cnn')
plt.ylim(0, 5)
plt.savefig('loss_policy_res_cnn.png')
plt.show()
```

```python

loss_cnn = np.load('loss_cnn.npy')
print(loss_cnn.shape)
print(loss_cnn)

x = [i / loss_cnn[0] for i in loss_cnn]
#draw the figure
plt.figure()
plt.plot(x)
plt.xlabel('iteration')
plt.ylabel('loss')
plt.title('loss of policy cnn')
plt.ylim(0, 20)
plt.savefig('loss_policy_cnn.png')
plt.show()
```

```python
# plot the loss of policy mlp, policy res_cnn, policy cnn
plt.figure()
plt.plot(loss_mlp[:40] / 100000000, label='mlp')
plt.plot(loss_res_cnn[:40] / 100000000, label='res_cnn', color='yellow')
plt.plot(loss_cnn[:40] / 100000000, label='cnn', color='red')
plt.xlabel('iteration')
plt.ylabel('loss')
plt.title('loss of mlp, res_cnn, cnn')
plt.ylim(0, 2)
plt.savefig('loss_policy.png')
plt.legend()
```
